```python
import pandas as pd
import os
import warnings
warnings.simplefilter("ignore")

os.getcwd()

# READS WHATEVER EXCEL FILE IS IN THE LOCAL FOLDER
#df = pd.read_excel("ToEdit.xlsx")
df = pd.read_excel("May.xlsx")


# REMOVES USELESS 1st COLUMN
df = df.iloc[: , 1:]


# OBTAINS ALL ROWS THAT ARE SAMPLES AND SUMS THE TOTAL

df_samples = df.loc[(df['Unit Price'] == 0.00) | (df['Unit Price'] == 0.01) | (df['Unit Price'] == 0.02) | (df['Unit Price'] == 0.03)
                   | (df['Unit Price'] == 0.10) | (df['Unit Price'] == 1.00)]

sample_total = df_samples['Line Total Price'].sum()
print('Sample total: $',sample_total)

# LOOKS FOR ANY COLLABS - GETS QUANTITY AND TOTAL SUM

print("------ Collab Totals ------")

df_houseplant = df.loc[df['Item'].str.contains('Houseplant')]
houseplant_qty = df_houseplant['Order Qty'].sum()
houseplant_ttl = df_houseplant['Line Total Price'].sum()
print('Houseplant Quantity:', houseplant_qty)
print('Houseplant Total: $', houseplant_ttl)

df_wham = df.loc[df['Item'].str.contains('Wham')]
wham_qty = df_wham['Order Qty'].sum()
wham_ttl = df_wham['Line Total Price'].sum()
print('WHAM Quantity:', wham_qty)
print('WHAM Total: $', wham_ttl)

df_zabores = df.loc[df['Item'].str.contains('Zabores')]
zabores_qty = df_zabores['Order Qty'].sum()
zabores_ttl = df_zabores['Line Total Price'].sum()
print('Zabores Quantity:', zabores_qty)
print('Zabores Total: $', zabores_ttl)

total_collabs = houseplant_ttl + wham_ttl + zabores_ttl
print('Total Collabs: $', total_collabs)

# DROPS ALL ROWS THAT HAVE IRRATIONAL RACCOON AS DBA
df_nosamples = df.drop(df.index[(df['Unit Price'] == 0.00) | (df['Unit Price'] == 0.01) | (df['Unit Price'] == 0.02) | (df['Unit Price'] == 0.03)
                   | (df['Unit Price'] == 0.10) | (df['Unit Price'] == 1.00)])
df_nosamples.drop(df_nosamples.index[df_nosamples['Destination DBA'] == 'IRRATIONAL RACCOON LLC'], inplace=True)

# GETS SALES REP TOTALS

df_reps = df_nosamples[df_nosamples["Item"].str.contains("Houseplant|Wham|Zabores") == False]

print("------ Sales Rep Totals ------")
df_alex = df_reps.loc[df_reps['Sales Rep'].str.contains('Alex Raffio')]
alex = df_alex['Line Total Price'].sum()
print('Alex Total Sales: $', alex)

df_carlos = df_reps.loc[df_reps['Sales Rep'].str.contains('Carlos Dardon')]
carlos = df_carlos['Line Total Price'].sum()
print('Carlos Total Sales: $', carlos)

df_doug = df_reps.loc[df_reps['Sales Rep'].str.contains('Doug Dixon')]
doug = df_doug['Line Total Price'].sum()
print('Doug Total Sales: $', doug)

df_house = df_reps.loc[df_reps['Sales Rep'].str.contains('House Account THCD')]
house = df_house['Line Total Price'].sum()
print('House Total Sales: $', house)

df_jairia = df_reps.loc[df_reps['Sales Rep'].str.contains('Jairia Pass')]
jairia = df_jairia['Line Total Price'].sum()
print('Jairia Total Sales: $', jairia)

df_payam = df_reps.loc[df_reps['Sales Rep'].str.contains('Payam Mousavinia')]
payam = df_payam['Line Total Price'].sum()
print('Payam Total Sales: $', payam)

df_brett = df_reps.loc[df_reps['Sales Rep'].str.contains('Brett Morales')]
brett = df_brett['Line Total Price'].sum()
print('Brett Total Sales: $', brett)

df_amanda = df_reps.loc[df_reps['Sales Rep'].str.contains('Amanda')]
amanda = df_amanda['Line Total Price'].sum()
print('Amanda Total Sales: $', amanda)

# GETS SKU COUNTS

print("------ SKU Counts ------")

df_dd14 = df_reps.loc[df_reps['Item'].str.contains('14g')]
dd14 = df_dd14['Order Qty'].sum()
print("Day Dreams 14g: ", dd14)

df_dd7 = df_reps.loc[df_reps['Item'].str.contains('7.0g')]
dd7 = df_dd7['Order Qty'].sum()
print("Day Dreams 7g: ", dd7)

df_dd3 = df_reps.loc[df_reps['Item'].str.contains('Day Dreams') & (df['Item'].str.contains('3.5g'))]
dd3 = df_dd3['Order Qty'].sum()
print("Day Dreams 3.5g: ", dd3)

df_infused = df_reps.loc[df_reps['Item'].str.contains('Infused')]
ddinfused = df_infused['Order Qty'].sum()
print("Dreamers Infused Prerolls: ", ddinfused)

df_eaze = df_reps.loc[df_reps['Item'].str.contains('EAZE ONLY')]
ddeaze = df_eaze['Order Qty'].sum()
print("EAZE ONLY: ", ddeaze)

df_pre = df_reps.loc[df_reps['Item'].str.contains('THCD Pre-Roll')]
ddpre = df_pre['Order Qty'].sum()
print("1g Prerolls:", ddpre)

df_6p = df_reps.loc[df_reps['Item'].str.contains('Tins')]
dd6p = df_6p['Order Qty'].sum()
print("6 Pack Prerolls:", dd6p)

df_pouch = df_reps.loc[df_reps['Item'].str.contains('Pouch')]
ddpouch = df_pouch['Order Qty'].sum()
print("1g Pouches", ddpouch)

df_jars = df_reps.loc[df_reps['Item'].str.contains('Jar')]
ddjars = df_jars['Order Qty'].sum()
jars = ddjars - ddeaze
print("Eighth Jars:", jars)

df_calirise = df_reps.loc[df_reps['Item'].str.contains('Cali Rise')]
ddcali = df_calirise['Order Qty'].sum()
print("Cali Rise:", ddcali)

# SAVES THE MODIFIED FILE TO ANOTHER EXCEL FILE 
df_reps.to_excel('testy.xlsx', index = False)

```
